```python
import pandas as pd
import  numpy as np
#loading the data set and setting the tittle of collumns
header_names = ['LBE','LB','AC','FM','UC','ASTV','MSTV',
                'ALTV','MLTV','DL','DS','DP','DR','Width',
                'Min','Max','Nmax','Nzeros','Mode','Mean',
                'Median','Varience','Tendency','NSP']

df = pd.read_csv('cardiotocography.csv/cardiotocography.csv', names = header_names)
print(df.info())

df.head()
#display the first 5 columns
```

```python
import numpy as np
import pandas as  pd

df_nan = df.replace(['?'], np.nan)
print(df_nan.isna().sum())
df_nan.head()#replacing questionmarks with Nan
```

```python
#converting objects to ints and floats
int_features = ['LBE','LB','AC','FM','UC','ASTV',
                'ALTV','DL','DS','DP','DR','Width',
                'Min','Max','Nmax','Nzeros','Mode','Mean',
                'Median','Varience','Tendency','NSP']

string_features = ['MLTV','MSTV']


df_nan[string_features + int_features] = df_nan[string_features + int_features].apply(pd.to_numeric, downcast='float',errors='coerce')

#converting int features 
for feature_name in int_features:
    df_nan[feature_name] = pd.array(df_nan[feature_name].values,dtype=pd.Int64Dtype())
```

```python
#correlation
corr = df_nan.corr()
print(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,10))
sns.heatmap(data=corr, annot = True, linewidths=.5, fmt='.1f')
```

```python
#heatmap showing correlation
import seaborn as sns
plt.figure(figsize=(10, 10))
ax = sns.heatmap(
    corr,
    vmin=-1, vmax=1 , center=0,
    cmap=sns.diverging_palette(20,220,n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
ax
```
